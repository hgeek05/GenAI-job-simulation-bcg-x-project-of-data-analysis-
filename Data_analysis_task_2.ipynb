```python
import pandas as pd

# Load the data
df = pd.read_csv("last_3_Years_Apple_Tesla (1).csv", sep=';')
```

```python
import string

def simple_chatbot(user_query):
    processed_query = user_query.lower().translate(str.maketrans('', '', string.punctuation))

    if processed_query == "what is the total revenue":
        total = df['Total Revenue ($B)'].sum()
        return f"The total revenue across all companies and years is ${total:,} Billion."

    elif processed_query == "how has net income changed over the last year":
        apple = df[df['Company'] == 'Apple'].sort_values('Fiscal Year')
        net_income_change = apple['Net Income ($B)'].iloc[-1] - apple['Net Income ($B)'].iloc[-2]
        return f"Apple's net income changed by ${net_income_change:,} Billion from {apple['Fiscal Year'].iloc[-2]} to {apple['Fiscal Year'].iloc[-1]}."

    elif processed_query == "what are the total assets":
        assets_by_company = df[df['Fiscal Year'] == 2023].groupby('Company')['Total Assets ($B)'].sum()
        response = "Total assets in 2023:\n"
        for company, assets in assets_by_company.items():
            response += f"- {company}: ${assets:,} Billion\n"
        return response.strip()

    elif processed_query == "compare revenue of apple and tesla":
        apple_revenue = df[(df['Company'] == 'Apple') & (df['Fiscal Year'] == 2023)]['Total Revenue ($B)'].values[0]
        tesla_revenue = df[(df['Company'] == 'Tesla') & (df['Fiscal Year'] == 2023)]['Total Revenue ($B)'].values[0]
        return f"In 2023, Apple had ${apple_revenue:,} Billion in revenue, while Tesla had ${tesla_revenue:,} Billion."

    elif processed_query == "what is microsofts cash flow for 2023":
        ms_cashflow = df[(df['Company'] == 'Microsoft') & (df['Fiscal Year'] == 2023)]['Operating Cash Flow ($B)'].values[0]
        return f"Microsoft’s cash flow from operating activities in 2023 was ${ms_cashflow:,} Billion."


    else:
        return " Sorry, I can only respond to predefined financial questions."
```

```python
if __name__ == "__main__":
    print("Welcome to the Financial Chatbot ")
    print("Try asking one of the following:\n")
    print("- What is the total revenue?")
    print("- How has net income changed over the last year?")
    print("- What are the total assets?")
    print("- Compare revenue of Apple and Tesla.")
    print("- What is Microsoft’s cash flow for 2023?\n")

    while True:
        query = input("You: ")
        if query.lower() in ["exit", "quit"]:
            print("Goodbye! ")
            break
        response = simple_chatbot(query)
        print("Bot:", response, "\n")
```
